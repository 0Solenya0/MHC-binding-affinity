```python
from sklearn.cluster import SpectralClustering
import pandas as pd
from torch import nn
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('./data.csv')
```

```python
import pylcs

mhc = df.sort_values(by=['MHC_type'])['MHC_sequence'].unique().tolist()
N = len(mhc)
adj = np.zeros((N, N))

for i in range(N):
    for j in range(i, N):
        common = pylcs.lcs_sequence_length(mhc[i], mhc[j])
        min_len = min(len(mhc[i]), len(mhc[j]))
        adj[i][j] = common / min_len
        adj[j][i] = common / min_len
```

```python
c = SpectralClustering(n_clusters=10, affinity='precomputed', assign_labels='cluster_qr', random_state=0).fit_predict(adj)
```

```python
for i in range(10):
    for j in range(c.shape[0]):
        if c[j] == i:
            print(df[df['MHC_sequence'] == mhc[j]]['MHC_type'].tolist()[0])
    print('-------------')
```
